```python
import numpy as np
import pandas as pd
```

```python
training_data = pd.read_csv('../_02_06_python_np_pd_plt/storepurchasedata.csv')
training_data.describe()
```

```python
X = training_data.iloc[:, :-1].values
y = training_data.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size =.20,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# minkowski is for ecledian distance
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
```

```python
# Model training
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
y_prob = classifier.predict_proba(X_test)[:,1]
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```

```python
new_prediction = classifier.predict(sc.transform(np.array([[40,20000]])))
new_prediction_proba = classifier.predict_proba(sc.transform(np.array([[40,20000]])))[:,1]
print(f"Prediction: \t{new_prediction}, \tprobability: \t{new_prediction_proba}")
```

```python
new_pred = classifier.predict(sc.transform(np.array([[42,50000]])))
new_pred_proba = classifier.predict_proba(sc.transform(np.array([[42,50000]])))[:,1]
print(f"Prediction: \t{new_pred}, \tprobability: \t{new_pred_proba}")
```
